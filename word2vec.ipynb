# Importing text data from wikipedia using webscraping method

```python
url = 'https://en.wikipedia.org/wiki/Machine_learning'

#getting the data from the url
html_data = requests.get(url).text

#using beautiful soup to convert the data from html format to readable format
beautiful_soup = BeautifulSoup(html_data, 'html5lib')

#finding all the paragarph tags to get data
text = beautiful_soup.find_all('p')

#removing unwanted spaces and replacing new lines
for word in text:
  print(word.text.strip().replace('\n', ''))
```

```python
for word in text:
  #appending the extracted information from url to file using file operations
  with open('/content/drive/MyDrive/Colab-Notebooks/Word2vec/machine_learning2.txt','a') as writing:
    writing.write(word.text.strip().replace('\n',''))
```

```python
#verifying the file which is written in the above step using file operations
with open('/content/drive/MyDrive/Colab-Notebooks/Word2vec/machine_learning2.txt','r')as readfile:
  print(readfile.read())
```

```python
#function to tokenize the words
def tokenizer(text):
  #empty list to write the file contents as tokens
  text_data_list = []
  #opening file in read mode
  with open(text,'r') as readfile:
    #appending list by removing unnecessary characters from textfile
    text_data_list.append(re.sub('[\W_]+',' ',readfile.read()))
    #splitting the list into tokens
    text_data_list = text_data_list[0].split()
  return text_data_list

#input for the tokenizer function
tokenized_data = tokenizer('/content/drive/MyDrive/Colab-Notebooks/Word2vec/machine_learning2.txt')


#verifying the tokenized_data
print(tokenized_data)

#dict to find the occurance of each word
word_counts = {}

for i,words in enumerate(tokenized_data):
  #appending words and their occurance in the tokenized data list
  if words in word_counts:
    word_counts[words] += 1
  else:
    word_counts[words]=1

print(word_counts)
```
